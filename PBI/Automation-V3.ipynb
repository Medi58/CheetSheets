```python
# Vandalism

df_FMEA_van=df_FMEA_dup[df_FMEA_dup['TCS_CAUSE']== 'Vandalism or Willful Damage  '] # Vandalism
df_FMEA_van_de= df_FMEA_van[df_FMEA_van['WO_DESCRIPTION']== 'TCS-CE: Earth                           ']
df_FMEA_van_de_earth= df_FMEA_van_de[df_FMEA_van_de['customer_resource'].str.contains('earth')]# down earth 


df_FMEA_van_de_earth
```

```python
df_FMEA_van_de_earth= df_FMEA_van_de[df_FMEA_van_de['customer_resource'].str.contains('earth')] # filter for down earth

# modify values 
df_FMEA_van_de_earth['Area']=df_FMEA_van_de_earth['Area'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'), 'Structures')
df_FMEA_van_de_earth['Asset Type']=df_FMEA_van_de_earth['Asset Type'] .mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'),'Pole - Component')
df_FMEA_van_de_earth['Asset Component']=df_FMEA_van_de_earth['Asset Component'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'),'Down Earth')
df_FMEA_van_de_earth['OBSERVED STATE (W0)']=df_FMEA_van_de_earth['OBSERVED STATE (W0)'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'),   'Missing')
df_FMEA_van_de_earth['FAILURE MECHANISM (W1)']=df_FMEA_van_de_earth['FAILURE MECHANISM (W1)'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'), 'Malicious Act')
df_FMEA_van_de_earth['IMMEDIATE CAUSE (W2)']=df_FMEA_van_de_earth['IMMEDIATE CAUSE (W2)'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'),   'Vandalism')
df_FMEA_van_de_earth['UNDERLYING CAUSE (W3)']=df_FMEA_van_de_earth['UNDERLYING CAUSE (W3)'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'), 'N/A')
df_FMEA_van_de_earth['FAILURE CLASSIFICATION']=df_FMEA_van_de_earth['FAILURE CLASSIFICATION'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'), 'Partial Failure')
df_FMEA_van_de_earth['VALIDATOR COMMENT']=df_FMEA_van_de_earth['VALIDATOR COMMENT'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'),    'Robot: Stolen Earth')
df_FMEA_van_de_earth['STATUS']=df_FMEA_van_de_earth['STATUS'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'),      'SIGNED OFF' )  
df_FMEA_van_de_earth['INITIAL']=df_FMEA_van_de_earth['INITIAL'].mask(df_FMEA_van_de_earth['customer_resource'].str.contains('stolen'),      'ROBOT' )  
```
